#1. What is hypothesis testing in statistics*

- Hypothesis testing is a statistical method used to determine whether there is enough evidence to reject a presumed assumption (hypothesis) about a population parameter.



#2 What is the null hypothesis, and how does it differ from the alternative hypothesis*

- Null Hypothesis (H‚ÇÄ): Assumes no effect or no difference in the population.

- Alternative Hypothesis (H‚ÇÅ): Suggests a significant effect or difference exists.

#3. What is the significance level in hypothesis testing, and why is it important.

 The probability of rejecting H‚ÇÄ when it is actually true (Type 1 error).

- Common values: 0.05 (5%) or 0.01 (1%).

- A lower Œ± reduces the likelihood of a Type 1 error but increases the risk of  
 a Type 2 error

#4. What does a P-value represent in hypothesis testing.

The probability of obtaining a test result at least as extreme as the observed result, assuming H‚ÇÄ is true.

- A small P-value (‚â§ Œ±) suggests strong evidence against H‚ÇÄ, leading to rejection of H‚ÇÄ.

- A large P-value (> Œ±) suggests insufficient evidence to reject H‚ÇÄ.


#5.  How do you interpret the P-value in hypothesis testing*

- P ‚â§ Œ±: Reject H‚ÇÄ (statistically significant result).

- P > Œ±: Fail to reject H‚ÇÄ (not statistically significant).
For example, if P = 0.03 and Œ± = 0.05, we reject H‚ÇÄ and conclude that there is a statistically significant effect.

#6. What are Type 1 and Type 2 errors in hypothesis testing.

- Type 1 Error (False Positive): Rejecting a true H‚ÇÄ. Probability = Œ±.

- Type 2 Error (False Negative): Failing to reject a false H‚ÇÄ. Probability = Œ≤.
Reducing one error type increases the risk of the other.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing.

- One-Tailed Test: Tests for an effect in one specific direction (e.g., mean is greater than a value).

- Two-Tailed Test: Tests for an effect in both directions (e.g., mean is different from a value).
A two-tailed test is used when the direction of the effect is unknown.

#10. What is the T-distribution, and when should it be used instead of the normal distribution.

- The sample size is small (n < 30).

- The population standard deviation (œÉ) is unknown.

```python
#1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import scipy.stats as stats
import numpy as np

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.2f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (significant difference).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (no significant difference).")

# Example Usage
sample_data = [52, 48, 50, 47, 53, 51, 49, 50, 54, 46]  # Sample data
population_mean = 50  # Population mean
population_std = 3  # Population standard deviation

z_test(sample_data, population_mean, population_std)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.2f}")
    print(f"P-Value: {p_value:.4f}")

    # Decision Rule
    z_critical = stats.norm.ppf(1 - alpha/2)  # Critical value for two-tailed test
    reject_region = abs(z_score) > z_critical

    if reject_region:
        print("Conclusion: Reject the null hypothesis (significant difference).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (no significant difference).")

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, color='blue', label="Standard Normal Distribution")

    # Shade critical regions
    x_critical_right = np.linspace(z_critical, 4, 100)
    y_critical_right = stats.norm.pdf(x_critical_right, 0, 1)
    plt.fill_between(x_critical_right, y_critical_right, color='red', alpha=0.5, label="Rejection Region")

    x_critical_left = np.linspace(-4, -z_critical, 100)
    y_critical_left = stats.norm.pdf(x_critical_left, 0, 1)
    plt.fill_between(x_critical_left, y_critical_left, color='red', alpha=0.5)

    # Plot the Z-score
    plt.axvline(x=z_score, color='black', linestyle='dashed', label=f"Z-Score: {z_score:.2f}")

    plt.axvline(x=z_critical, color='red', linestyle='dotted', label=f"Critical Z: ¬±{z_critical:.2f}")
    plt.axvline(x=-z_critical, color='red', linestyle='dotted')

    plt.title("Two-Tailed Z-Test Decision Regions")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

# Example: Generate random sample data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std=5, Size=30
population_mean = 50
population_std = 5

# Perform two-tailed Z-test
two_tailed_z_test(sample_data, population_mean, population_std)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(mu_null, mu_alt, sigma, alpha=0.05, n=30):
    x = np.linspace(mu_null - 4*sigma/np.sqrt(n), mu_alt + 4*sigma/np.sqrt(n), 1000)
    null_dist = stats.norm.pdf(x, mu_null, sigma/np.sqrt(n))
    alt_dist = stats.norm.pdf(x, mu_alt, sigma/np.sqrt(n))

    z_critical = stats.norm.ppf(1 - alpha)
    critical_value = mu_null + z_critical * (sigma / np.sqrt(n))

    plt.figure(figsize=(8, 5))
    plt.plot(x, null_dist, label="Null Distribution (H‚ÇÄ)", color='blue')
    plt.plot(x, alt_dist, label="Alternative Distribution (H‚ÇÅ)", color='green')

    x_type1 = np.linspace(critical_value, mu_null + 4*sigma/np.sqrt(n), 100)
    y_type1 = stats.norm.pdf(x_type1, mu_null, sigma/np.sqrt(n))
    plt.fill_between(x_type1, y_type1, color='red', alpha=0.5, label="Type 1 Error (Œ±)")

    x_type2 = np.linspace(mu_null - 4*sigma/np.sqrt(n), critical_value, 100)
    y_type2 = stats.norm.pdf(x_type2, mu_alt, sigma/np.sqrt(n))
    plt.fill_between(x_type2, y_type2, color='orange', alpha=0.5, label="Type 2 Error (Œ≤)")

    plt.axvline(critical_value, color='black', linestyle='dashed', label=f"Critical Value: {critical_value:.2f}")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 1 and Type 2 Errors")
    plt.legend()
    plt.show()

visualize_type1_type2_errors(mu_null=50, mu_alt=53, sigma=5, alpha=0.05, n=30)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The means are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between means.")

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=10, size=30)

independent_t_test(sample1, sample2)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_rel(sample1, sample2)

    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between paired samples.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between paired samples.")

    plt.figure(figsize=(6, 4))
    plt.scatter(sample1, sample2, color='blue', label="Paired Data Points")
    plt.plot([min(sample1), max(sample1)], [min(sample2), max(sample2)], 'r--', label="y = x Line")
    plt.xlabel("Before Treatment")
    plt.ylabel("After Treatment")
    plt.title("Paired Sample Comparison")
    plt.legend()
    plt.show()

np.random.seed(42)
before_treatment = np.random.normal(loc=50, scale=5, size=30)
after_treatment = before_treatment + np.random.normal(loc=2, scale=3, size=30)

paired_t_test(before_treatment, after_treatment)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)

    margin_of_error = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1) * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"{confidence*100:.1f}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
    return lower_bound, upper_bound

np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)

confidence_interval(sample)
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, likelihood, marginal_likelihood):
    posterior = (likelihood * prior) / marginal_likelihood
    print(f"Posterior Probability: {posterior:.4f}")
    return posterior

prior_prob = 0.02
likelihood_given_evidence = 0.85
marginal_likelihood = (likelihood_given_evidence * prior_prob) + ((1 - prior_prob) * 0.10)

bayes_theorem(prior_prob, likelihood_given_evidence, marginal_likelihood)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed, expected):
    chi2, p_value = stats.chisquare(observed, expected)

    print(f"Chi-Square Statistic: {chi2:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
    else:
        print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

observed_data = np.array([50, 30, 20])
expected_data = np.array([40, 40, 20])

goodness_of_fit_test(observed_data, expected_data)

```

```python
#16.  Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    f_statistic = var1 / var2 if var1 > var2 else var2 / var1
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

    print(f"F-Statistic: {f_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in variances.")

np.random.seed(42)
sample1 = np.random.normal(50, 10, 30)
sample2 = np.random.normal(50, 15, 30)

f_test(sample1, sample2)
```

```python
#17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

import numpy as np
import scipy.stats as stats

def anova_test(*groups):
    f_statistic, p_value = stats.f_oneway(*groups)

    print(f"F-Statistic: {f_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject the null hypothesis: At least one group mean is significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between group means.")

np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

anova_test(group1, group2, group3)


```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    print("Checking Assumptions for ANOVA:\n")

    # Normality Test (Shapiro-Wilk Test)
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1} - Shapiro-Wilk Test: p = {p_value:.4f} {'(Fail to reject normality)' if p_value > 0.05 else '(Reject normality)'}")

    # Homogeneity of Variance Test (Levene‚Äôs Test)
    stat, p_value = stats.levene(*groups)
    print(f"\nLevene‚Äôs Test for Equal Variance: p = {p_value:.4f} {'(Fail to reject equal variance)' if p_value > 0.05 else '(Reject equal variance)'}")

    # Boxplot for Visual Inspection
    plt.boxplot(groups, labels=[f"Group {i+1}" for i in range(len(groups))])
    plt.title("Boxplot of Groups")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.show()

np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

check_anova_assumptions(group1, group2, group3)
```

```python
#21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_f_distribution(dfn, dfd):
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, dfn, dfd)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'F-Distribution (df1={dfn}, df2={dfd})', color='blue')
    plt.fill_between(x, y, alpha=0.3, color='blue')
    plt.axvline(stats.f.ppf(0.95, dfn, dfd), color='red', linestyle='dashed', label="Critical Value (Œ±=0.05)")

    plt.xlabel("F-Statistic")
    plt.ylabel("Probability Density")
    plt.title("F-Distribution Visualization")
    plt.legend()
    plt.show()

visualize_f_distribution(dfn=5, dfd=10)
```

```python
#24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
import scipy.stats as stats

def variance_hypothesis_test(sample_size=30, sample_variance=20, population_variance=25, alpha=0.05):
    chi_square_stat = (sample_size - 1) * sample_variance / population_variance
    lower_critical = stats.chi2.ppf(alpha / 2, df=sample_size - 1)
    upper_critical = stats.chi2.ppf(1 - alpha / 2, df=sample_size - 1)
    p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df=sample_size - 1),
                       1 - stats.chi2.cdf(chi_square_stat, df=sample_size - 1))

    print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
    print(f"Critical Values: ({lower_critical:.4f}, {upper_critical:.4f})")
    print(f"P-Value: {p_value:.4f}")

    if chi_square_stat < lower_critical or chi_square_stat > upper_critical:
        print("Result: Reject the null hypothesis (Variance is significantly different).")
    else:
        print("Result: Fail to reject the null hypothesis (Variance is not significantly different).")

np.random.seed(42)
sample = np.random.normal(50, np.sqrt(20), 30)
sample_variance = np.var(sample, ddof=1)

variance_hypothesis_test(sample_size=30, sample_variance=sample_variance, population_variance=25, alpha=0.05)
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
import scipy.stats as stats

def z_test_proportions(success_a, size_a, success_b, size_b, alpha=0.05):
    p1 = success_a / size_a
    p2 = success_b / size_b
    pooled_p = (success_a + success_b) / (size_a + size_b)
    standard_error = np.sqrt(pooled_p * (1 - pooled_p) * (1/size_a + 1/size_b))

    z_statistic = (p1 - p2) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    print(f"Proportion 1: {p1:.4f}, Proportion 2: {p2:.4f}")
    print(f"Z-Statistic: {z_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Result: Reject the null hypothesis (Significant difference between proportions).")
    else:
        print("Result: Fail to reject the null hypothesis (No significant difference between proportions).")

z_test_proportions(success_a=45, size_a=100, success_b=30, size_b=100)
```
